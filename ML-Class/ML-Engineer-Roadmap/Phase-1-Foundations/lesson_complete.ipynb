## 1. Python Core : Structures de données

**Pourquoi ?** ML = manipulation listes grandes échelles → besoin structures efficaces.

- **Listes** : Mutable, indexées.
- **Dictionnaires** : Key-value (ex: features → valeurs).

```python
# Listes et comprehensions (fast list filter)
villes = ['Dakar', 'Thies', 'Dakar']
collectes = [25.5, 12.3, None]

# Filtre non-null (list comprehension)
collectes_clean = [q for q in collectes if q is not None]
print('Clean:', collectes_clean)

# Dict pour metadata
dataset_info = {'rows': 8, 'features': ['ville', 'quantite', 'collecte'], 'theme': 'dechets'}
print(dataset_info)
```

## 2. NumPy : Fondations numériques (Vectorisation)

**Maths** : Array = matrice 1D/N-D. Ops vectorisées O(n) vs boucles O(n^2).

$\mu = \frac{1}{n} \sum x_i$ (mean)

**Avantage ML** : Entraînement modèles = millions ops parallèles.

```python
import numpy as np

# Array création
quantites_np = np.array([25.5, 18.0, 30.1, 12.3])
print('Array:', quantites_np)
print('Type:', quantites_np.dtype)

# Stats descriptives
print(f"Moyenne: {quantites_np.mean():.2f} kg")
print(f"Écart-type: {quantites_np.std():.2f} kg")  # Dispersion
print(f"Median: {np.median(quantites_np):.2f} kg")

# Vectorisation (fast)
normalise = (quantites_np - quantites_np.mean()) / quantites_np.std()
print('Normalisé:', normalise)
```

## 3. Pandas : Manipulation tabulaire (DataFrame = Excel++)

**Structure** : Rows (observations), Columns (features).

**Méthodes clés :**
| Méthode | But | Exemple |
|---------|-----|---------|
| `read_csv` | Load | pd.read_csv('fichier.csv') |
| `head()` | Aperçu | 5 premières lignes |
| `isnull()` | Missing values | Bool mask |
| `fillna()` | Imputer | Remplacer NaN |
| `query()` | Filtre SQL-like | df.query("ville == 'Dakar'") |

**Data types** : float (quantite), object (ville), bool (collecte).

```python
import pandas as pd

# 1. Load data
df = pd.read_csv('sample_data.csv')
print("Aperçu :")
print(df.head())
print(f"Forme : {df.shape} (lignes, colonnes)")
print("\nTypes :")
print(df.dtypes)
```

```python
# 2. EDA (Exploratory Data Analysis)
print("Stats descriptives :")
print(df.describe())  # Numériques seulement

print("\nValeurs manquantes :")
print(df.isnull().sum())
```

```python
# 3. Nettoyage (Preprocessing crucial pour ML)
# Stratégie : mean imputation pour quantite
moyenne_q = df['quantite'].mean()
df['quantite'] = df['quantite'].fillna(moyenne_q)
print("Après fillna :")
print(df.isnull().sum())
print(df.head())
```

```python
# 4. Filtrage / Sélection
# Dakar ET quantite > 20
dakar_gros = df.query("ville == 'Dakar' and quantite > 20")
print("Dakar gros déchets :")
print(dakar_gros)

# Group by
moy_ville = df.groupby('ville')['quantite'].mean()
print("\nMoyenne par ville :")
print(moy_ville)
```

## 4. Visualisation : Matplotlib (Insights visuels)

**Règle d'or** : Graph = 1000 stats.

Types :
- Bar : Catégoriques (villes).
- Hist : Distributions.
- Scatter : Corrélations.

```python
import matplotlib.pyplot as plt

# Bar plot : Moyenne par ville
moy_ville.plot(kind='bar')
plt.title('Quantité moyenne déchets par ville')
plt.ylabel('kg')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram : Distribution quantités
plt.figure(figsize=(8,5))
plt.hist(df['quantite'], bins=6, edgecolor='black', alpha=0.7)
plt.title('Distribution quantités déchets')
plt.xlabel('Quantité (kg)')
plt.ylabel('Fréquence')
plt.grid(True, alpha=0.3)
plt.show()

print("Insight : La plupart <20kg → normal pour collecte.")
```
