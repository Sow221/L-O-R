## 3. Pandas - Roi des DataFrames

```python
import pandas as pd

# Lire CSV
df = pd.read_csv('../sample_data.csv')
print(df.head())
print(df.shape)
```

```python
# Nettoyer valeurs manquantes
print('Missing:', df.isnull().sum())
df_clean = df.fillna(0)
print(df_clean.head())
```
